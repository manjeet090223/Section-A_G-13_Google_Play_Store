# 1. Environmental Configuration

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Google-Playstore-Cleaned.csv', encoding='latin1', low_memory=False)
```

## 2. KPI Feature Engineering

```python
df['Is_Successful'] = (df['Rating'] >= 4.0).astype(int)


install_threshold = df['Installs_Numeric'].quantile(0.75)
df['High_Volume_App'] = (df['Installs_Numeric'] >= install_threshold).astype(int)


df['Price_Segment'] = pd.cut(df['Price'],
                             bins=[-1, 0, 5, 20, float('inf')],
                             labels=['Free', 'Budget (<$5)', 'Mid-Range', 'Premium'])
```

# 3. Date Normalization for Time-Series Dashboards

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')

df['Release_Year'] = df['Released'].dt.year
df['Update_Month'] = df['Last Updated'].dt.month_name()
```

# 4. Data Integrity Check (Pre-Visualization)

```python
print("Final Columns for Tableau Import:")
print(df[['Is_Successful', 'High_Volume_App', 'Price_Segment', 'Release_Year']].isnull().sum())

df.head()
```

# 5. Final Export for Tableau/Power BI

```python
df.to_csv('Google_Playstore_Final_Load.csv', index=False, encoding='utf-8')

print("Notebook 05 Complete: Final dataset is ready for Dashboarding.")
```
